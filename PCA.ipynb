```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
% matplotlib inline         
```

```python
from sklearn.datasets import load_breast_cancer
bf=load_breast_cancer()
bf
```

```python
print(bf['DESCR'])
print(bf['target_names'])
```

```python
df=pd.DataFrame(bf['data'],columns=bf['feature_names'])
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(df)
s=scaler.transform(df)
s
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca.fit(s)
```

```python
x=pca.transform(s)
print(s.shape)
print(x.shape)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x[:,0],x[:,1],c=bf['target'],cmap='plasma')
plt.xlabel("first principle component")
plt.ylabel("second principle component")
```

```python
pca.components_
```

```python
df_comp=pd.DataFrame(pca.components_,columns=bf['feature_names'])
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap='plasma')
```
